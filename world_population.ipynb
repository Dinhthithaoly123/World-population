Dân số thế giới tiếp tục tăng khoảng 140 người mỗi phút, với tỷ lệ sinh nhiều hơn tử ở hầu hết các quốc gia. Tuy nhiên, nhìn chung, tốc độ tăng dân số đã chậm lại trong vài thập kỷ. Sự chậm lại này dự kiến sẽ tiếp tục cho đến khi tốc độ tăng dân số bằng 0 (số ca sinh và số ca tử vong bằng nhau). Hiện nay, hai quốc gia tỷ dân chiếm tới 1/3 dân số thế giới là Trung Quốc và Ấn Độ. Cả hai nước này đều có dân số trên 1,4 tỷ người. Dân số thế giới hiện nay là hơn 8 tỷ người, với 140 triệu trẻ em được sinh ra mỗi năm. 

```python
import pandas as pd
import numpy as np
import seaborn as sns
import  matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("D:/Projects/world_population.csv")
print(df.head(10))
```

```python
label_encoder = LabelEncoder()
df['Country/Territory']  = label_encoder.fit_transform(df['Country/Territory'])
df['CCA3'] = label_encoder.fit_transform(df['CCA3'])
df['Capital'] =label_encoder.fit_transform(df['Capital'])
df['Continent'] =  label_encoder.fit_transform(df['Continent'])
print(df.head(10))
```

### Phân tích biến động dân số theo thời gian

```python
#Phân tích biến động dân số 5 quốc gia  bất kì theo thời gian
plt.figure(figsize = (10,6))
selected_countries = df['Country/Territory'].sample(n=5, random_state=1).tolist()
for country in selected_countries:
    country_data = df[df['Country/Territory'] == country]
    if not country_data.empty: 
        years = country_data.columns[6:14]
        population = country_data.iloc[:, 6:14].values.ravel()  # Lấy giá trị dân số từ các cột tương ứng
        plt.plot(years, population, label=country)

plt.title('Population Dynamics Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(loc='upper left', fontsize='small', bbox_to_anchor=(1, 1))
plt.grid(True)
plt.show()
```

### So sánh mức độ tăng trưởng dân số giữa các lục địa

```python
# Tạo một DataFrame con chứa thông tin về mức độ tăng trưởng dân số của các lục địa
continent_growth = df.groupby('Continent')['Growth Rate'].mean().reset_index()
# Biểu đồ
plt.figure(figsize=(10, 6))
bars = plt.bar(continent_growth['Continent'], continent_growth['Growth Rate'], color='skyblue')
plt.title('Average Population Growth Rate by Continent')
plt.xlabel('Continent')
plt.ylabel('Population Growth Rate')
plt.xticks(range(6), ['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America'])
plt.grid(axis='y')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 4), ha='center', va='bottom')
plt.tight_layout()
plt.show()
```

### Đánh giá mức độ tập trung dân số

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Density (per km²)'], df['2022 Population'], alpha=0.5)
plt.title('Population Density vs Population in 2022')
plt.xlabel('Density (per km²)')
plt.ylabel('Population')
plt.grid(True)
plt.show()
```

### Sự gia tăng dân số từ 1970 đến 2022

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Country/Territory'], df['2022 Population'] / df['1970 Population'], marker='o', linestyle='-')
plt.title('Population Growth Rate since 1970')
plt.xlabel('Country/Territory')
plt.ylabel('Population Growth Rate')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

### Liên kết dân số với diện tích và mật độ dân số

```python
# Tính mật độ dân số
df['Population Density'] = df['2022 Population'] / df['Area (km²)']

# Vẽ biểu đồ phân tán
plt.figure(figsize=(10, 6))
plt.scatter(df['Area (km²)'], df['2022 Population'], c=df['Population Density'], cmap='viridis', edgecolor='k', s=200)
plt.title('Population and Land Area')
plt.xlabel('Land Area (km²)')
plt.ylabel('Population')
plt.colorbar(label='Population Density')
plt.grid(True)
plt.show()
```

### Dự đoán dân số cho năm 2050

```python
# Tỷ lệ tăng trưởng dân số trung bình trong giai đoạn 1970-2022
df['Average Growth Rate'] = (df['2022 Population'] / df['1970 Population']) ** (1 / 52) - 1
# Dự đoán dân số tương lai cho năm 2050 (khoảng cách 28 năm kể từ 2022)
df['Predicted Population (2050)'] = df['2022 Population'] * (1 + df['Average Growth Rate']) ** 28

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
plt.bar(df['Continent'], df['Predicted Population (2050)'], color='skyblue')
plt.title('Predicted Population in 2050')
plt.xlabel('Continent')
plt.ylabel('Population')
plt.xticks(range(6), ['Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America'])
plt.grid(axis='y')
plt.show()
```
